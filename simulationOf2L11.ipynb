# Simulation of 2.L.1.1 Mystery RC

```python
%matplotlib inline

import matplotlib
import numpy as np
import scipy
import matplotlib.pyplot as plt
from compare_results import compare_results
from chi2fit import linear_fit
```

```python
# ohms and farads
C = 0.1e-6
R = 39e3
Rtest = 39e3
CwithPhase = np.complex64(-1j)/C
```

```python
freqs = np.logspace(1,4) # hertz
omega = 2*np.pi*freqs # rad/s
Xc = -1j/(omega*C)
```

```python
V1 = Xc+R+Rtest
V2 = Xc+R
phaseV1 = np.angle(V1)*180/np.pi
phaseV2 = np.angle(V2)*180/np.pi
ampV1 = np.abs(V1)
ampV2 = np.abs(V2)
```

```python
plt.semilogx(freqs,phaseV1,label="V1")
plt.semilogx(freqs,phaseV2,label="V2")
plt.semilogx(freqs,phaseV2-phaseV1,label="V2-V1")
plt.semilogx(freqs,np.angle(Xc)*180/np.pi,label="Xc")
plt.axvline(1/(2*np.pi*(R)*C),label="f(3 dB)",c='k')
plt.legend()
plt.xlabel("f (Hz)")
plt.ylabel("$\phi$ [deg]")
plt.title(r"Mystery R=39k$\Omega$ C=0.1$\mu$F R$_{test}=50$k$\Omega$")
```

```python
plt.loglog(freqs,ampV1,label="V1")
plt.loglog(freqs,ampV2,label="V2")
plt.loglog(freqs,np.abs(Xc),label="Xc")
#plt.loglog(freqs,ampV2-phaseV1,label="V2-V1")
plt.axvline(1/(2*np.pi*R*C),label="f(3 dB)",c="k")
plt.legend()
plt.xlabel("f (Hz)")
plt.ylabel("$|V|$ [arbitrary units]")
plt.title(r"Mystery R=39k$\Omega$ C=0.1$\mu$F R$_{test}=50$k$\Omega$")
```
